```python
def is_terminal(board, index):
    #print(board)
    SUM = 15
    #row and col length
    n = 3
    
    # index of last move taken
    #checks each row, col, diagonals for terminal move
    
    #check col
    colNum = index[1]
    colSum = 0
    for i in range(n):
        #if an entry is empty
        if(np.isinf(board[i][colNum])):
            break;
        else:
            colSum = colSum + board[i][colNum]
    if colSum == SUM:
        return True
    
    #check row
    rowNum = index[0]
    rowSum = 0
    for i in range(n):
        if(np.isinf(board[rowNum][i] == -1)):
            break;
        else:
            rowSum = rowSum + board[rowNum][i]
    if rowSum == SUM:
        return True
    
    #check diag
    diagSum = 0
    if(rowNum == colNum):
        diagSum = 0
        for i in range(n):
            if(np.isinf(board[i][i])):
                break;
            else:
                diagSum = diagSum + board[i][i]
    if diagSum == SUM:
        return True
            
    #check other diagonal
    
    if(rowNum + colNum == n - 1):
        for i in range(n):
            if(np.isinf(board[i][(n-1)-i])):
                break;
            else:
                diagSum = diagSum + board[i][i]
    if diagSum == SUM:
        return True
    
    return False
```
